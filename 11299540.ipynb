### Cleaning Data::
<hr>

```python
wine.duplicated()
```

```python
wine.drop_duplicates( inplace=True)
wine.info()
```

```python
column_list= list(wine.columns)
for column in column_list:
    if pd.api.types.is_numeric_dtype(wine[column]):
        decimal_places = wine[column].apply(lambda x: len(str(x).split('.')[1]) if '.' in str(x) else 0).max()
        c_mean=wine[column].mean()
        c_mean=round(c_mean,decimal_places)
        wine[column].fillna(c_mean ,inplace= True )
wine.info()
wine
```

### Visualizations--Analysis Using Graphs
<hr>

#### 1. This is a scatter plot of fixed acidity against volatile acidity


```python
plt.scatter(wine['fixed acidity'],wine['volatile acidity'],alpha=0.5,color= 'red',)
plt.title('Fixed Acidity vs. Volatile Acidity')
plt.xlabel('Fixed Acidity')
plt.ylabel('Volatile Acidity')
plt.grid(True)
plt.show()
```

<p>The graph above shows a visual representation of the fixed acidity of red wine against its volatile acidity properties.<br>
    The correlation between the two properties is calculated below:
</p>

```python
correlation = wine['fixed acidity'].corr(wine['volatile acidity'])
f"Correlation between Fixed Acidity and Volatile Acidity: {correlation}"
```

<p>The correlation is -0.25 (in 2 decimal places).<br>This means that for a slight increase in the value of fixed acidity, there will be a decrease in the volatile acidity and vice versa</p>

#### 2. This is a pie chart to aid in analysing the quality of the red wine from the population of samples in the dataset

```python
quality_counts=wine["quality"].value_counts()
quality_counts
```

```python
plt.figure(figsize=(9, 9))
plt.pie(quality_counts, startangle=90)
plt.legend(quality_counts.index, title='Quality', loc='upper right')
plt.title('Distribution of Wine Quality')
plt.show()
```

```python
x=list(quality_counts)
y=list()
for i in x:
    y.append(round(100*(i/1360),2))
    
y
```

<p>
    The analysis of the 1360 samples in terms of percentages is given by the list 'y' above. 
    While the pie chart gives a visual representation.
    It can be deduced from the chart and from 'y' that::</p>

    - About 43% of the samples have a quality of 5.
    - About 40% have a quality of 6.
    - About 12% have a quality of 7.
